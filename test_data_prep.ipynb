```python
import pandas as pd
import json
```

```python
df = pd.read_json("2025-November.json")
```

```python
df
```

```python
df['question'] = df['llm_chat_history'].apply(lambda x: x[0]['content'] if isinstance(x, list) and len(x) > 0 and isinstance(x[0], dict) and 'content' in x[0] else None)
```

```python
df
```

```python
# Prepare JSONL file with thread_id as id and question
jsonl_data = df[['thread_id', 'question']].rename(columns={'thread_id': 'id'})

# Export to JSONL format
with open('output.jsonl', 'w', encoding='utf-8') as f:
    for _, row in jsonl_data.iterrows():
        json.dump({'id': row['id'], 'question': row['question']}, f, ensure_ascii=False)
        f.write('\n')
```
